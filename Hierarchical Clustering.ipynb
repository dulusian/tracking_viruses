```python
from sklearn.cluster import KMeans
import csv
import numpy as np
from numpy import genfromtxt
import sys
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster import hierarchy
from scipy.cluster.hierarchy import dendrogram, linkage
import csv
```

```python
def stripCSVTOMatrix(file):
    my_data = genfromtxt(file, delimiter=',')
    my_data = my_data[1:,1:]
    return my_data
```

```python
def makeAdjacencyMatrix(m):
    np.fill_diagonal(m,0)
```

```python
def deleteBelowThreshold(m, threshold):
    threshold_indices = m <= threshold
    m[threshold_indices] = 0
    m = m.reshape(3168,3168)
    return m
```

```python
def make_matrix_from_corona_csv(csv_file):
    species = []
    geoloc = {}
    host = {}
    with open(csv_file) as csv_file:
        list_file = list(csv.reader(csv_file, delimiter=','))
        line_count = 0
        for row in list_file[1:]:
            species.append(row[9])
            geoloc[line_count] = row[18]
            host[line_count] = row[19]
            print(f'Species - {row[9]}, GeoLoc - {row[18]}, Host - {row[19]}')
            line_count += 1
        print(f'Processed {line_count} lines.')
    return list_file, species, geoloc, host
```

```python
def unique_list(l1):
    return np.unique(l1)
```

```python
def species_with_ac_dict(species, acronym):
    return_dict = {}
    for i in range(len(species)):
        return_dict[acronym[i]] = species[i]
    return return_dict
```

```python
def make_species_acronym(l1):
    return_list = []
    for x in l1:
        acronym = "".join(e[0] for e in x.split())
        return_list.append(acronym.upper())
    return return_list
```

```python
def make_host_acronym(l1):
    return_list = []
    for x in l1:
        acronym = "_".join(e[0:2] for e in x.split())
        return_list.append(acronym.upper())
    return return_list
```

```python
def country_ac_names(csv_file):
    country_ac_name_dict = {}
    with open(csv_file) as csv_file:
        list_file = list(csv.reader(csv_file, delimiter=','))
        line_count = 0
        for row in list_file:
            country_ac_name_dict[row[1]] = row[0]
    return country_ac_name_dict
```

```python
def split_string_until_char(l1):
    return [r.split(":")[0] for r in l1]
```

```python
def change_country_to_ac(l1,country_dict):
    return_list = []
    for l in l1:
        if l == "":
            return_list.append("")
        else:
            return_list.append(country_dict[l])
    return return_list
```
